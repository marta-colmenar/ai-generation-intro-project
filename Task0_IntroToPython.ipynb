# Introduction to Python

Here is a brief introduction to some fundamental concepts in Python programming. Let's see what you have learned!

We will cover:

1. Variables and data types
2. Lists and loops
3. Functions
4. Conditionals (if/else)
5. A simple hands-on exercise


### Variables and Data Types

```python
# Numbers
x = 5
y = 3
z = x + y
print("x + y =", z)
```

```python
# Strings
name = "Alice"
greeting = "Hello " + name
print(greeting)
```

```python
# Boolean
is_happy = True
print("Is happy?", is_happy)
```

### Lists and Loops

```python
# List of numbers
numbers = [1, 2, 3, 4, 5]

# Loop through the list
for n in numbers:
    print(f"{n} squared is {n**2}")
```

```python
# Loop through a list of names
names = ["Alice", "Bob", "Charlie"]
for name in names:
    print("Hello", name)
```

### Functions

```python
# Simple function
def add(a, b):
    return a + b

print("10 + 5 =", add(10, 5))
```

```python
# Function with condition
def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False

print("Is 4 even?", is_even(4))
print("Is 7 even?", is_even(7))
```

### Conditionals (if/else)

```python
# Simple if/else
age = 16

if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
```
