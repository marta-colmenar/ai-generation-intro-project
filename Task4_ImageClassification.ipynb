Most of these models are trained on **huge datasets with millions of images**, which helps them learn to identify a wide variety of objects. One of the most famous datasets is **ImageNet**, which contains over 14 million labeled images across more than 20,000 categories.

### Image Classification in Python

```python
from transformers import pipeline
from PIL import Image
from IPython.display import display

# Load image classifier
classifier = pipeline("image-classification", model="google/vit-base-patch16-224")
```

```python
# Function to classify local images
def classify_image_local(file_path):
    image = Image.open(file_path)
    import matplotlib.pyplot as plt

    FIGSIZE = (4, 4)
    fig = plt.figure(figsize=FIGSIZE)
    plt.imshow(image)
    plt.axis('off')
    display(fig)
    plt.close(fig)
    results = classifier(image)
    print("Predictions:")
    for r in results:
        print(f"  {r['label']} — {r['score']*100:.2f}%")
    print("-"*40)
```

```python
# Test the images
classify_image_local("images/dog.jpeg")
classify_image_local("images/cat.jpeg")
classify_image_local("images/banana.jpeg")
```
